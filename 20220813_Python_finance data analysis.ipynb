# Pandas基礎入門

## 資料載入

Pandas常用的資料類型有：
1. Series：一維陣列的資料，列(index)為資料儲存順序
2. DataFrame：二維的資料陣列，由列(index)及欄(column)組合而成

Pandas模組匯入

```
import pandas as pd
```

## Series 一維資料

```
se = pd.Series(串列)
```

範例資料：

```
stocks = ['聯電', '台積電', '聯發科', '大力光', '鴻海']
price = [42, 510, 694, 2115, 109]
```

````python
# 宣告
import pandas as pd

# 一維資料 Series
price = [42, 510, 694, 2115, 109]
se = pd.Series(price)
se
# 求值
se.values

# 求索引
se.index
# list(se.index)

# 自訂index
stocks = ['聯電', '台積電', '聯發科', '大立光', '鴻海']
price = [42, 510, 694, 2115, 109]
se2 = pd.Series(price, index=stocks)
se2
````

````python
# 資料資訊 shape, describe()
se2.shape
se2.describe()
````

````python
# 統計：加總、最小值、最大值、平均值、中位數
se2.sum()
se2.min()
se2.max()
se2.mean()
se2.median()
````

````python
se2
````

````python
# Series 取值
# 用索引值(排的順序, 由0開始)
se2[1]  # 台積電
se2[3]  # 大立光

# iloc(用索引值), loc(用索引名稱)
se2.iloc[1]
se2.iloc[3]
se2.loc['台積電']
se2.loc['大立光']

# 一次取多個值
se2.iloc[[1, 3, 4]]
se2.loc[["台積電", "大立光", "鴻海"]]

# 一次取多個連續值, iloc用range()的觀念
se2.iloc[1:4]
se2.loc["台積電":"大立光"]
````

````python
se.iloc[1]
se.loc[1]
````

## 資料連接與合併

* concat()：水平或垂直連接
* merge()：合併

````python
# 統一發票中獎號碼
# 本期 https://invoice.etax.nat.gov.tw/index.html
# 上期 https://invoice.etax.nat.gov.tw/lastNumber.html

import pandas as pd
url1 = 'https://invoice.etax.nat.gov.tw/index.html'
url2 = 'https://invoice.etax.nat.gov.tw/lastNumber.html'
df1 = pd.read_html(url1)[0]
df2 = pd.read_html(url2)[0]
````

````python
df2
````

````python
# concat()垂直連接
df3 = pd.concat([df1, df2], ignore_index=True)
df3
````

````python
# concat()水平連接
df4 = pd.concat([df1, df2], axis=1)
df4
````

````python
# 資料整理：標題、去除最後一列
# df1.columns = ['獎別', '0506月中獎號碼']
# df2.columns = ['獎別', '0304月中獎號碼']
df1.drop(df1.index[-1], axis=0, inplace=True)
df2.drop(df2.index[-1], axis=0, inplace=True)
````

````python
# merge()相同欄位合併
df5 = pd.merge(df1, df2)
df5
````
